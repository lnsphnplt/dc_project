# Calculate Amino acid make UP

Because this takes a huge amount of data we need change the database format. I found no sql database which can handle more than 8000 columns.

```python
import os
import pandas as pd
import sqlite3

if os.getenv("SNAKEMAKE_EXECUTION"):
    # Adjust paths for Snakemake execution
    db_path = "./02-Data_Curation/unified-curated.db"
else:
    # Adjust paths for local execution
    db_path = "../02-Data_Curation/unified-curated.db"

con = sqlite3.connect(db_path)

df = pd.DataFrame(pd.read_sql_query("select seq from prod", con))
df.head()

```

# Makeup calculation

```python
from propy import PyPro

def calculate_make_up(seq):
    pro = PyPro.GetProDes(seq)
    res = pro.GetAAComp()
    res.update(pro.GetDPComp())
    res.update(pro.GetTPComp())
    return res


len(calculate_make_up(df['seq'].iloc[0]))
```
